```python
import numpy as np

def get_student_names(num_students):
    names = []
    for i in range(num_students):
        name = input(f"Enter name for student {i+1}: ")
        names.append(name)
    return names

def get_marks(num_students, num_subjects, names):
    marks_array = np.zeros((num_students, num_subjects), dtype=float)

    for i in range(num_students):
        print(f"--- Entering marks for {names[i]} ---")
        for j in range(num_subjects):
            while True:
                try:
                    marks = float(input(f"Input marks for subject {j+1} for {names[i]}: "))
                    if 0 <= marks <= 100:
                        marks_array[i, j] = marks
                        break
                    else:
                        print("  Marks should be between 0 and 100. Please try again.")
                except ValueError:
                    print("  Invalid input. Please enter a number.")
    return marks_array

def calculate_grade(percentage):
    if percentage >= 90:
        return 'A+'
    elif percentage >= 80:
        return 'A'
    elif percentage >= 70:
        return 'B+'
    elif percentage >= 60:
        return 'B'
    elif percentage >= 50:
        return 'C'
    else:
        return 'F'

def display_results(names, total_marks, percentages, grades):
    """Display results in a formatted table"""
    print("\n" + "="*80)
    print(f"{'Student Name':<20} {'Total Marks':<15} {'Percentage':<15} {'Grade':<10}")
    print("-"*80)

    for i in range(len(names)):
        print(f"{names[i]:<20} {total_marks[i]:<15.2f} {percentages[i]:<15.2f} {grades[i]:<10}")

    print("="*80)

def main():
    print()


    while True:
        try:
            num_students = int(input("Enter number of students: "))
            num_subjects = int(input("Enter number of subjects: "))
            if num_students > 0 and num_subjects > 0:
                break
            else:
                print("Please enter positive numbers.")
        except ValueError:
            print("Invalid input. Please enter integers.")


    names = get_student_names(num_students)


    marks = get_marks(num_students, num_subjects, names)


    total_marks = np.sum(marks, axis=1)


    max_possible_marks = num_subjects * 100
    percentages = (total_marks / max_possible_marks) * 100


    grades = np.array([calculate_grade(p) for p in percentages])


    display_results(names, total_marks, percentages, grades)





if __name__ == "__main__":
    main()
```
